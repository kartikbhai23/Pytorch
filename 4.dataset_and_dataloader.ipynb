# PyTorch Datasets and DataLoaders
This notebook demonstrates how to design a custom PyTorch dataset and use `DataLoader` to feed batch inputs to our models. Custom datasets and dataloaders are crucial for handling large out-of-memory files, image preprocessing, and efficient batching.

We will cover:
1. Synthesizing data using `scikit-learn`.
2. Wrapping custom features and labels in PyTorch's `Dataset` class.
3. Using `DataLoader` to handle batching, shuffling, and multi-threaded loading.

## 1. Importing Libraries
We import the classification data generator from `sklearn` alongside the core PyTorch modules.

```python
# Importing libraries
import sklearn
from sklearn.datasets import make_classification
import torch
```

## 2. Generating Synthetic Classification Data
We create a small classification problem containing 10 samples and 2 informative features.

```python
# Step 1: Creating dataset
x, y = make_classification(
    n_samples=10,        # Number of samples
    n_features=2,        # Number of features
    n_informative=2,     # Number of informative features
    n_redundant=0,       # Number of redundant features
    n_classes=2,         # Number of binary target classes
    random_state=42,     # Seed for reproducibility
)
```

## 3. Inspecting the Generated Data

```python
# Display the feature matrix
x
```

```python
# Display shape of feature matrix
x.shape
```

## 4. Converting Numpy Arrays to PyTorch Tensors
PyTorch modules operate on Tensors. We convert the features (`x`) and labels (`y`) into float32 tensors.

```python
# Converting the dataset arrays to PyTorch Tensors
# Fix: Crucial correction to load labels 'y' instead of overwriting with features 'x'
x = torch.tensor(x, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32)
```

## 5. Designing the Custom Dataset Class
To build a custom PyTorch dataset, we inherit from `torch.utils.data.Dataset` and implement:
- `__init__`: Standard constructor where we load features and labels.
- `__len__`: Returns the total count of samples in the dataset.
- `__getitem__`: Returns the feature and corresponding label for a given sample index.

```python
from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        # Return total samples
        return self.features.shape[0]
    
    def __getitem__(self, index):
        # Return single sample tuple (features, label)
        return self.features[index], self.labels[index]
```

## 6. Instantiating and Testing the Custom Dataset

```python
# Instantiate dataset wrapper
dataset = CustomDataset(x, y)
```

```python
# Get total length of dataset
len(dataset)
```

```python
# Fetch the 3rd sample (index 2)
dataset[2]
```

## 7. Creating the PyTorch DataLoader
The `DataLoader` takes care of grouping samples into mini-batches, shuffling the dataset at each epoch, and handling multi-threaded processing.

```python
# Create DataLoader with batch size 2 and shuffle enabled
dataLoader = DataLoader(
    dataset,
    batch_size=2,
    shuffle=True
)
```

## 8. Iterating through Batches
We iterate through the loader to verify that data is loaded in shuffled batches of shape `(batch_size, num_features)`.

```python
# Print batch features and labels to check shuffling and batch dimensions
for batch_features, batch_labels in dataLoader:
    print("Batch Features Shape:", batch_features.shape)
    print("Batch Labels:", batch_labels)
    print("Batch Features:\n", batch_features)
    print("-" * 50)
```
