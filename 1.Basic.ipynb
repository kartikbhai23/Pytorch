## 6. Matrix Multiplication
Matrix multiplication is the fundamental build block of linear layer transformations. In PyTorch, this is computed using `torch.matmul` or the `@` operator.


```python
f = torch.randint(size =(2,3), low =0, high =10)
g = torch.randint(size =(3,2), low =0, high =10)

print(f)
print(g)
```

```python
#Matrix multiplication
torch.matmul(f,g)
```

```python
# Initialize a PyTorch tensor
vect1 = torch.tensor([1,2])
vect2 = torch.tensor([3,4])
```

```python
#dot prodcut
torch.dot(vect1,vect2)
```

```python
#transpose
torch.transpose(f,0,1)
```

## 9. In-place Operations
Operators postfixed with an underscore (e.g., `add_()`, `sub_()`) modify the tensor's contents directly inside its current memory address.


```python
m = torch.rand(2,3)
n = torch.rand(2,3)

print(m)
print(n)
```

```python
# to store addition on same variable
# ( - )
m + n  # it store sum in new tensor 
```

```python
m.add_(n) # it store value in m tensor
```

```python
torch.relu(m) # it stor value in new tensor
```

```python
m.relu_() # it store value in m tensor
```

```python
b=a.numpy() # convert tensor to numpy array 
```
